### Outlier

```python
df.head()
```

```python
non_menu_cols = ['ì˜ì—…ì¼ì', 'holiday']
op_cols = [col for col in df.columns if col.endswith('_ìš´ì˜')]
non_menu_cols.extend(op_cols)
menu_columns = [col for col in df.columns if col not in non_menu_cols]
```

```python
outlier_info = {}
```

```python
for menu in menu_columns:
    series = df[menu]
    
    sales_data = series[series > 0]

    if sales_data.count() < 4:
        continue
    
    Q1 = sales_data.quantile(0.25)
    Q3 = sales_data.quantile(0.75)
    IQR = Q3 - Q1
    
    if IQR == 0:
        continue
     
    upper_bound = Q3 + 1.5 * IQR
   
    outliers = df[(df[menu] > upper_bound)]
 
    if not outliers.empty:
        outlier_info[menu] = {
            'ìƒí•œì„ ': round(upper_bound, 2),
            'ë°œê²¬ëœ ì´ìƒì¹˜': outliers[['ì˜ì—…ì¼ì', menu]].to_dict('records')
        }
        df[menu] = series.clip(upper=upper_bound)

print("\n ì´ìƒì¹˜ ì¡°ì •ì´ ì™„ë£Œë˜ì—ˆìŠµë‹ˆë‹¤.")
```

```python
print("\n--- ì´ìƒì¹˜ ì¡°ì • ë¦¬í¬íŠ¸ ---")
if not outlier_info:
    print("ëª¨ë“  ë©”ë‰´ì—ì„œ IQR ê¸°ì¤€ì„ ë²—ì–´ë‚˜ëŠ” ì´ìƒì¹˜ê°€ ë°œê²¬ë˜ì§€ ì•Šì•˜ìŠµë‹ˆë‹¤.")
else:
    for menu, info in outlier_info.items():
        print(f"\n ë©”ë‰´: {menu}")
        print(f"  - IQR ìƒí•œì„ : {info['ìƒí•œì„ ']}")
        print(f"  - ì¡°ì •ëœ ë°ì´í„° ìˆ˜: {len(info['ë°œê²¬ëœ ì´ìƒì¹˜'])}ê±´")
        for record in info['ë°œê²¬ëœ ì´ìƒì¹˜']:
            date = record['ì˜ì—…ì¼ì'].strftime('%Y-%m-%d')
            original_value = record[menu]
            print(f"    - {date}: {original_value:.2f} -> {info['ìƒí•œì„ ']:.2f} ìœ¼ë¡œ ì¡°ì •")
print("--------------------------")
```

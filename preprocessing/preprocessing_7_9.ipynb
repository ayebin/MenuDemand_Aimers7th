### 운영

```python
menu_columns = [col for col in df.columns if col not in ['영업일자', 'holiday']]
store_names = sorted(list(set([col.split('_')[0] for col in menu_columns])))
print(store_names)
```

```python
for store in store_names:
    store_menu_columns = [col for col in menu_columns if col.startswith(store + '_')]
    store_daily_sales = df[store_menu_columns].sum(axis=1)
    rolling_sales_sum = store_daily_sales.rolling(window=7).sum()
    is_operating = pd.Series(1, index=df.index)
    is_operating[rolling_sales_sum == 0] = 0
    df[f'{store}_운영'] = is_operating
```

### 주중 판매

```python
day_of_week = df['영업일자'].dt.dayofweek
```

```python
non_menu_cols = ['영업일자', 'holiday']

op_cols = [col for col in df.columns if col.endswith('_운영')]
non_menu_cols.extend(op_cols)
```

```python
all_menu_columns = [col for col in df.columns if col not in non_menu_cols]
store_names = sorted(list(set([col.split('_')[0] for col in all_menu_columns])))
```

```python
analysis_results = {}
```

```python
for store in store_names:
    analysis_results[store] = {'주중 전용 메뉴': [], '주말 전용 메뉴': []}
    store_menu_columns = [col for col in all_menu_columns if col.startswith(store + '_')]
    for menu in store_menu_columns:
        total_weekday_sales = df.loc[day_of_week < 5, menu].sum()
        total_weekend_sales = df.loc[day_of_week >= 5, menu].sum()
        if total_weekday_sales > 0 and total_weekend_sales == 0:
            analysis_results[store]['주중 전용 메뉴'].append(menu)
        if total_weekend_sales > 0 and total_weekday_sales == 0:
            analysis_results[store]['주말 전용 메뉴'].append(menu)
```

```python
for store, results in analysis_results.items():
    print(f"\n매장: {store}")
    print("-" * 20)

    if results['주중 전용 메뉴']:
        print("  [☀️ 주중에만 판매하는 메뉴]")
        for menu_name in results['주중 전용 메뉴']:
            print(f"    - {menu_name}")
    else:
        print("  - 주중에만 판매하는 메뉴는 없습니다.")

    if results['주말 전용 메뉴']:
        print("  [주말에만 판매하는 메뉴]")
        for menu_name in results['주말 전용 메뉴']:
            print(f"    - {menu_name}")
    else:
        print("  - 주말에만 판매하는 메뉴는 없습니다.")
    print("-" * 20)
```

### Outlier

```python
df.head()
```

```python
non_menu_cols = ['영업일자', 'holiday']
op_cols = [col for col in df.columns if col.endswith('_운영')]
non_menu_cols.extend(op_cols)
menu_columns = [col for col in df.columns if col not in non_menu_cols]
```

```python
outlier_info = {}
```

```python
for menu in menu_columns:
    series = df[menu]
    
    sales_data = series[series > 0]

    if sales_data.count() < 4:
        continue
    
    Q1 = sales_data.quantile(0.25)
    Q3 = sales_data.quantile(0.75)
    IQR = Q3 - Q1
    
    if IQR == 0:
        continue
     
    upper_bound = Q3 + 1.5 * IQR
   
    outliers = df[(df[menu] > upper_bound)]
 
    if not outliers.empty:
        outlier_info[menu] = {
            '상한선': round(upper_bound, 2),
            '발견된 이상치': outliers[['영업일자', menu]].to_dict('records')
        }
        df[menu] = series.clip(upper=upper_bound)

print("\n 이상치 조정이 완료되었습니다.")
```

```python
print("\n--- 이상치 조정 리포트 ---")
if not outlier_info:
    print("모든 메뉴에서 IQR 기준을 벗어나는 이상치가 발견되지 않았습니다.")
else:
    for menu, info in outlier_info.items():
        print(f"\n 메뉴: {menu}")
        print(f"  - IQR 상한선: {info['상한선']}")
        print(f"  - 조정된 데이터 수: {len(info['발견된 이상치'])}건")
        for record in info['발견된 이상치']:
            date = record['영업일자'].strftime('%Y-%m-%d')
            original_value = record[menu]
            print(f"    - {date}: {original_value:.2f} -> {info['상한선']:.2f} 으로 조정")
print("--------------------------")
```
